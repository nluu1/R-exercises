## Creating Functions for Visualization

```python
def myplot(playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c='Black', ls='--', marker='s', ms=10, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```

```python
myplot(["KobeBryant", "LeBronJames", "DerrickRose"])
```

```python
#--------- More functions design: adding color customization for each item with for loop
```

```python
# To generate color for each of the item in for loop, we create a dictionary for the variations

def myplot(playerlist):
    Col = {"KobeBryant":"Black", "LeBronJames":"Red", "DerrickRose":"Blue","JoeJohnson":"Green","CarmeloAnthony":"Magenta","DwightHoward":"Black","ChrisBosh":"Red","ChrisPaul":"Blue","KevinDurant":"Green","DwayneWade":"Magenta"}
    Mrk = {"KobeBryant":"s", "LeBronJames":"o", "DerrickRose":"^","JoeJohnson":"D","CarmeloAnthony":"s","DwightHoward":"o","ChrisBosh":"^","ChrisPaul":"D","KevinDurant":"s","DwayneWade":"o"}
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c=Col[name], ls='--', marker=Mrk[name], ms=7, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
      
```

```python
myplot(["KobeBryant","JoeJohnson","LeBronJames","CarmeloAnthony","DwightHoward","ChrisBosh","ChrisPaul","KevinDurant","DerrickRose","DwayneWade"])
```

```python
#-- Add to function to locate another type of data group
def myplot(data,playerlist = Players): #---- add a default parameters
    Col = {"KobeBryant":"Black", "LeBronJames":"Red", "DerrickRose":"Blue","JoeJohnson":"Green","CarmeloAnthony":"Magenta","DwightHoward":"Black","ChrisBosh":"Red","ChrisPaul":"Blue","KevinDurant":"Green","DwayneWade":"Magenta"}
    Mrk = {"KobeBryant":"s", "LeBronJames":"o", "DerrickRose":"^","JoeJohnson":"D","CarmeloAnthony":"s","DwightHoward":"o","ChrisBosh":"^","ChrisPaul":"D","KevinDurant":"s","DwayneWade":"o"}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Col[name], ls='--', marker=Mrk[name], ms=7, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
      

myplot(Points, ["KobeBryant", "LeBronJames", "DerrickRose"])
```

```python
myplot(Salary) #if no player specified, default (Players) will be visualized
```
